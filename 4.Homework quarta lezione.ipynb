```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
```

```python
C = pd.read_csv('Cereal.csv')

C_sC = C.sort_values('calories')  
C_sF = C.sort_values('fat')
C_sS = C.sort_values('sugars')
C_sB = C.sort_values('carbo')
C_sP = C.sort_values('protein')

C.head()
```

```python
#Calorie nei cereali

plt.figure(figsize=(10, 5))                           
plt.bar('name', 'calories', data = C, color = 'black');
```

```python
#Dal meno calorico a salire

plt.figure(figsize=(20, 10))                  
plt.bar('name', 'calories', data = C_sC, color = 'Teal')
plt.xlabel('Si dovrebbero leggere i nomi delle marche', size = 25, color = 'Green')
plt.ylabel('Calorie', size = 30, color = 'Green');
```

```python
#Ha un p pi di senso 

plt.figure(figsize=(10, 17), facecolor = '0.75')                          
plt.barh('name', 'calories', data = C_sC, color = 'Teal', edgecolor = 'Black', linewidth = 1)
plt.xlabel('Calorie', size = 25)
plt.title('Classifica Cereali Calorici', size = 30);
```

```python
fig,ax = plt.subplots()   
ax.plot(C.name, C_sF.fat, label = 'Grassi')
ax.plot(C.name, C_sS.sugars, label = 'Zuccheri')
ax.plot(C.name, C_sP.protein, label = 'Proteine')
ax2 = ax.twinx()
ax2.plot(C.name, C_sC.calories, label = 'Calorie', color = 'Red')
ax2.set_ylabel('Calorie', size = 25) 
ax.set_ylabel('Quantit Macro/hg', size = 25) 
ax.set_xlabel('Marche di Cereali', size = 25)
ax.set_title("Quantit calorica rispetto macronutrienti", size = 30)

#per vare due legende diverse ma in zone differenti
#ax.legend(fontsize = 'xx-large')
#ax2.legend(fontsize = 'xx-large', loc="right")

#fig.legend(loc="upper right") ma la legenda rimane fuori dal grafico
fig.legend(loc="upper left", bbox_to_anchor=(0,1), bbox_transform=ax.transAxes, fontsize = 'xx-large')

fig.set_size_inches(15, 10);
```

```python
#Moltiplicando per 10 i valori dei macro altrimenti troppo piccoli da osservare (conservando quindi le proporzioni)
#si riesce meglio a visualizzare come,  mantendo una sola scala di valori per l'asse y, l'impennata delle calorie corrisponda 
#all'aumento dei grassi rispetto a quella degli altri macro

fig,ax = plt.subplots()   
ax.plot(C.name, C_sC.calories, label = 'Calorie')
ax.plot(C.name, C_sF.fat * 10, label = 'Grassi x10')
ax.plot(C.name, C_sS.sugars * 10, label = 'Zuccheri x10')
ax.plot(C.name, C_sP.protein *10, label = 'Proteine x10')
ax.set_ylabel('Quantit/hg', size = 25) 
ax.set_xlabel('Marche di Cereali', size = 25)
ax.set_title("Quantit calorica rispetto macronutrienti", size = 30)
plt.legend(fontsize = 'xx-large')
ax.annotate('Impennata', xy=(27, 10), xytext=(35, 16),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Impennata', xy=(30, 110), xytext=(35, 116),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Seconda Impennata', xy=(57, 19), xytext=(63, 10),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Seconda Impennata', xy=(61, 120), xytext=(63, 105),
            arrowprops=dict(facecolor='black', shrink=0.05))


fig.set_size_inches(15, 10)
plt.grid()
```

```python
#Si pu vedere con i carboidrati invece che con i soli zuccheri

fig,ax = plt.subplots()
ax.plot(C.name, C_sC.calories, label = 'Calorie')
ax.plot(C.name, C_sF.fat * 10, label = 'Grassi x10')
ax.plot(C.name, C_sB.carbo * 10, label = 'Carboidrati x10')
ax.plot(C.name, C_sP.protein *10, label = 'Proteine x10')
ax.set_ylabel('Quantit/hg', size = 25) 
ax.set_xlabel('Marche di Cereali', size = 25)
ax.set_title("Quantit calorica rispetto macronutrienti", size = 30)
plt.legend(fontsize = 'xx-large')
fig.set_size_inches(15, 10)
```
